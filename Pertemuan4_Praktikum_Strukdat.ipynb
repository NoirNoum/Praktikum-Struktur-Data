QUEUE DENGAN LIST

```python
# Python program to demonstrate queue implementation using list
queue = []

# append() function to enqueue element in the queue
queue.append('a')
queue.append('i')
queue.append('u')
queue.append('e')
queue.append('o')

print('initial queue')
print(queue)

# pop(0) function to dequeue element from queue in FIFO order
print ('\nElements dequeued from queue: ')
print(queue.pop(0))
print(queue.pop(0))
print(queue.pop(0))

print('\nQueue after removing elements')
print(queue)

# uncommenting print(queue.pop()) will cause an IndexError as the queue is now empty
```

Queue dengan collection.dequeue

```python
# Python program to demonsterate queue implementation using colllection.deque
from collections import deque

queue = deque()

# append() function to Enqueue element in the queue
queue.append('a')
queue.append('i')
queue.append('u')
queue.append('e')
queue.append('o')

print('Initail queue: ')
print(queue)

# popleft() function to dequeue element form stack in LIFO order
print("\nelements dequeue form the queue")
print(queue.popleft())
print(queue.popleft())
print(queue.popleft())

print("\nQueue after removing element")
print(queue)

# uncommenting print(queue.popleft()) will cause an IndexError  as the queue is now empty
```

Queue dengan queue.Queue

```python
# Python program to demonstrate queue implementation using queue module
from queue import Queue

# Initializing a queue
qu = Queue(maxsize=5)

# qsize() show the number of elements in the queue
print(qu.qsize())

# put() function to push element in the queue
qu.put('a')
qu.put('i')
qu.put('u')
qu.put('e')
qu.put('o')

print("Full: ", qu.full())
print("Size:", qu.qsize())

# get() function to pop element from queue in LIFO onder
print("\nElements popped from the queue")
print(qu.get())
print(qu.get())
print(qu.get())

print("\nEmpty: ", qu.empty())
print("Full: ", qu.full())
```

QUEUE DENGAN SINGLE LINKED LIST

```python
class Queue:
  def __init__(self):
    self.queue = list()

  def addtoqu(self,dataval):
# Insert method to add element
    if dataval not in self.queue:
      self.queue.insert(0, dataval)
      return True
    return False

# Pop method to remove element
  def removefromq(self):
    if len(self.queue)>0:
      return self.queue.pop()
    return ("No elements in Queue!")

  def size(self):
    return len(self.queue)

TheQueue = Queue()
print(TheQueue.addtoqu("Jan"))
print(TheQueue.addtoqu("Feb"))
print(TheQueue.addtoqu("Marc"))
print(TheQueue.addtoqu("April"))
print(TheQueue.size())
print(TheQueue.removefromq())
print(TheQueue.removefromq())
print(TheQueue.size())
```
