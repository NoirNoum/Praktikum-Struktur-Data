**Representasi Graf dengan Adjacency List**

Graf adalah struktur data yang terdiri dari simpul (vertex) dan sisi (edge). Dalam
Python, graf dapat direpresentasikan menggunakan dictionary yang berisi daftar
adjacency dari tiap simpul.

Contoh Kode Program:

```python
# Membuat graf tak berarah menggunakan adjacency list
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

# Menampilkan graf
def print_graph (graf) :
  for node in graf:
    print (f" {node} -> {graf [node] } ")
print_graph (graph)
```

**Penjelasan kode di atas:**

* Graf didefinisikan sebagai dictionary, di mana setiap kunci adalah simpul, dan
nilainya adalah list dari simpul-simpul yang terhubung.
* Fungsi print graph mencetak semua simpul dan daftar tetangganya.

**Traversal Graf menggunakan BFS**

Breadth-First Search (BFS) menjelajahi graf secara level demi level, dimulai dari simpul awal, kemudian mengunjungi simpul yang bertetangga, lalu simpul-simpul bertetangga dari tetangga, dan seterusnya.

Contoh Kode Program:


```python
from collections import deque

def bfs(graph, start):
    visited = set()
    queue = deque([start])

    while queue:
        vertex = queue.popleft()
        if vertex not in visited:
            print(vertex, end=" ")
            visited.add(vertex)
            queue.extend([neighbor for neighbor in graph[vertex] if neighbor not in visited])

# Contoh pemanggilan
bfs(graph, 'A')
```

**Penjelasan**
* deque digunakan untuk efisiensi antrian.
* Set visited menyimpan simpul yang telah dikunjungi.
* Fungsi mencetak urutan kunjungan dari simpul awal



---
**Traversal Graf menggunakan DFS**

Depth-First Search (DFS) menelusuri graf dengan cara menyelam sedalam mungkin ke satu jalur sebelum kembali dan menyusuri jalur lain.

Contoh kode program:

```python
def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()


    if start not in visited:
        print(start, end=" ")
        visited.add(start)
        for neighbor in graph[start]:
            dfs(graph, neighbor, visited)

# Contoh pemanggilan
dfs(graph, 'A')
```

**Penjelasan:**
* DFS menggunakan rekursi untuk menjelajahi simpul.
* Set visited mencegah kunjungan berulang.
* Fungsi mencetak simpul sesuai urutan DFS.

**1. Membuat Graph Baru dengan 5 Simpul dan 7 Sisi**

```python
graph_baru = {
    'A': ['B', 'C', 'D'],
    'B': ['A', 'C', 'E'],
    'C': ['A', 'B', 'D'],
    'D': ['A', 'C', 'E'],
    'E': ['B', 'D']
}
```

Penjelasan:

Graph di atas memiliki:

* 5 simpul = A, B, C, D, E
* 7 sisi penghubung

Graph tersebut termasuk undirected graph karena hubungan antar simpul bersifat dua arah.

**2. Implementasi BFS dari Simpul Berbeda**

```python
from collections import deque

# Fungsi BFS

def bfs(graph, start):
    visited = set()
    queue = deque([start])

    while queue:
        vertex = queue.popleft()

        if vertex not in visited:
            print(vertex, end=' ')
            visited.add(vertex)

            for neighbor in graph[vertex]:
                if neighbor not in visited:
                    queue.append(neighbor)

# Menjalankan BFS dari simpul C
bfs(graph_baru, 'C')
```

Penjelasan:

BFS mengunjungi simpul berdasarkan level terdekat terlebih dahulu.

**3. Implementasi DFS dari Simpul Berbeda**

```python
# Fungsi DFS

def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()

    if start not in visited:
        print(start, end=' ')
        visited.add(start)

        for neighbor in graph[start]:
            dfs(graph, neighbor, visited)

# Menjalankan DFS dari simpul B

dfs(graph_baru, 'B')
```

Penjelasan:

DFS akan menelusuri graph sedalam mungkin sebelum kembali ke jalur sebelumnya.

**4. Memodifikasi BFS agar Mengembalikan List**

```python
from collections import deque

# BFS dengan hasil list

def bfs_list(graph, start):
    visited = []
    queue = deque([start])

    while queue:
        vertex = queue.popleft()

        if vertex not in visited:
            visited.append(vertex)

            for neighbor in graph[vertex]:
                if neighbor not in visited:
                    queue.append(neighbor)

    return visited

print(bfs_list(graph_baru, 'A'))
```

Penjelasan:

Versi ini tidak hanya mencetak hasil traversal, tetapi juga menyimpan hasil traversal dalam bentuk list.

**5. Membuat Fungsi find_path()**

```python
# Fungsi mencari jalur

def find_path(graph, start, end, path=[]):
    path = path + [start]

    if start == end:
        return path

    for node in graph[start]:
        if node not in path:
            new_path = find_path(graph, node, end, path)

            if new_path:
                return new_path

    return None

print(find_path(graph_baru, 'A', 'E'))
```

Penjelasan:

Fungsi ini digunakan untuk mencari jalur dari satu simpul ke simpul tujuan.
