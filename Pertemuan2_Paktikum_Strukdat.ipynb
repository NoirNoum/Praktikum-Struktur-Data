LATIHAN 1: Double Linked List

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return

        current = self.head
        while current.next:
            current = current.next

        current.next = new_node
        new_node.prev = current

    def insert_middle(self, new_data):
        """Menyisipkan node baru tepat di tengah list"""
        if not self.head:
            self.head = Node(new_data)
            return

        # Hitung jumlah node
        count = 0
        current = self.head
        while current:
            count += 1
            current = current.next

        # Cari posisi tengah
        mid = count // 2
        current = self.head
        for _ in range(mid):
            current = current.next

        # Sisipkan sebelum node tengah saat ini
        new_node = Node(new_data)
        prev_node = current.prev

        new_node.next = current
        new_node.prev = prev_node

        if prev_node:
            prev_node.next = new_node
        else:
            self.head = new_node

        current.prev = new_node

    def delete_by_value(self, key):
        """Hapus node yang memiliki data == key"""
        current = self.head

        while current:
            if current.data == key:
                # Kasus 1: hanya 1 node
                if current == self.head and current.next is None:
                    self.head = None
                    return

                # Kasus 2: hapus head
                if current == self.head:
                    self.head = current.next
                    self.head.prev = None
                    return

                # Kasus 3: hapus tail
                if current.next is None:
                    current.prev.next = None
                    return

                # Kasus 4: hapus di tengah
                current.prev.next = current.next
                current.next.prev = current.prev
                return

            current = current.next

        print(f"Nilai '{key}' tidak ditemukan dalam list")

    def print_forward(self):
        if not self.head:
            print("List kosong")
            return

        current = self.head
        print("Forward: ", end="")
        while current:
            print(current.data, end=" <-> ")
            current = current.next
        print("None")

    def print_backward(self):
        if not self.head:
            return

        # Pergi ke node terakhir
        current = self.head
        while current.next:
            current = current.next

        print("Backward: ", end="")
        while current:
            print(current.data, end=" <-> ")
            current = current.prev
        print("None")


# Penggunaan
print("=== Doubly Linked List ===")
dll = DoublyLinkedList()

dll.append("A")
dll.append("B")
dll.append("C")

print("Awal:")
dll.print_forward()
dll.print_backward()

print("\nSetelah menyisipkan 'X' di tengah:")
dll.insert_middle("X")
dll.print_forward()

print("\nSetelah menghapus node dengan nilai 'B':")
dll.delete_by_value("B")
dll.print_forward()

print("\nSetelah menghapus node dengan nilai 'Z' (tidak ada):")
dll.delete_by_value("Z")
```

LATIHAN 2: Circular Linked List

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class CircularLinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)

        if not self.head:
            self.head = new_node
            new_node.next = self.head
            return

        current = self.head
        while current.next != self.head:
            current = current.next

        current.next = new_node
        new_node.next = self.head

    def append_end(self, data):
        """Sama dengan append (menyisipkan di akhir)"""
        self.append(data)

    def delete_end(self):
        """Hapus node terakhir (node sebelum kembali ke head)"""
        if not self.head:
            print("List sudah kosong")
            return

        # Hanya 1 node
        if self.head.next == self.head:
            self.head = None
            return

        # Cari node sebelum terakhir
        current = self.head
        prev = None

        while current.next != self.head:
            prev = current
            current = current.next

        # Hapus node terakhir
        prev.next = self.head
        # current akan dihapus oleh garbage collector

    def print_list(self):
        if not self.head:
            print("List kosong")
            return

        print("Circular List: ", end="")
        current = self.head
        while True:
            print(current.data, end=" -> ")
            current = current.next
            if current == self.head:
                break
        print("(back to head)")


# Penggunaan
print("\n=== Circular Linked List ===")
cll = CircularLinkedList()

cll.append("P")
cll.append("Q")
cll.append("R")

print("Awal:")
cll.print_list()

print("\nSetelah menyisipkan 'S' di akhir:")
cll.append_end("S")
cll.print_list()

print("\nSetelah menghapus node akhir:")
cll.delete_end()
cll.print_list()

print("\nSetelah menghapus node akhir lagi:")
cll.delete_end()
cll.print_list()
```
