#Case 1: Coin Change
**Masalah Coin Change & Pendekatan Greedy**

Masalah coin change bertujuan mencari jumlah koin minimum untuk mencapai target nilai tertentu menggunakan denominasi yang tersedia tanpa batas.

* Cara Kerja Greedy: Pada setiap langkah, algoritma selalu memilih koin dengan nilai terbesar yang muat dalam sisa target.

* Kelebihan & Batasan: Pendekatan ini sangat efisien dan akurat pada sistem koin standar (seperti 1, 5, 10, 25). Namun, pada sistem koin tertentu (misalnya 1, 3, 4), algoritma greedy bisa gagal memberikan solusi optimal (minimum).

Tujuan Studi Kasus: Melalui materi ini, mahasiswa akan belajar memahami mekanisme sekaligus mengenali batasan dari pendekatan greedy.

**Contoh Kode Program:**

```python
def coin_change_greedy(amount, coins) :
  coins.sort (reverse=True)
  result = []
  for coin in coins:
    while amount >= coin:
      amount -= coin
      result.append (coin)
  return result

# Contoh penggunaan
amount = 57
coins = [25, 10, 5, 1]
change = coin_change_greedy (amount, coins)
print ("Koin yang digunakan:", change)
```

**Penjelasan:**
* Fungsi coin_change_greedy menerima dua parameter: amount (jumlah uang
yang ingin dikembalikan) dan coins (daftar denominasi koin). Daftar koin
diurutkan dari terbesar ke terkecil untuk memastikan greedy memilih koin
bernilai tinggi terlebih dahulu. Dalam iterasi, selama amount masih lebih besar
atau sama dengan nilai koin, nilai koin tersebut dikurangkan dari amount dan
ditambahkan ke daftar hasil. Fungsi akhirnya mengembalikan daftar koin yang
digunakan.

# Case 2: Fractional Knapsack

**Masalah Fractional Knapsack (Ransel Pecahan)**

Fractional Knapsack adalah kasus klasik yang dapat diselesaikan secara optimal menggunakan algoritma greedy. Tujuannya adalah mengisi ransel berkapasitas tertentu dengan barang-barang (yang memiliki berat dan nilai) agar menghasilkan nilai total maksimal.

* Cara Kerja Greedy: Barang dipilih berdasarkan rasio nilai per unit berat tertinggi terlebih dahulu, lalu diambil sebanyak mungkin.

* Solusi Optimal: Karena barang boleh diambil dalam bentuk pecahan (sebagian), pendekatan greedy dijamin selalu menghasilkan solusi yang paling optimal.


**Contoh Kode Program:**

```python
def fractional_knapsack (items, capacity) :
  # items = list of tuples (value, weight)
  items = sorted(items, key=lambda x: x [0]/x[1], reverse=True)
  total_value = 0.0
  for value, weight in items:
    if capacity >= weight:
      capacity -= weight
      total_value += value
    else:
      total_value += value * (capacity / weight)
      break
  return total_value

# Contoh penggunaan
items = [(60, 10), (100, 20), (120, 30)] # (nilai, berat)
capacity = 50
max_value = fractional_knapsack (items, capacity)
print("Nilai maksimum yang dapat dibawa:", max_value)
```

**Penjelasan:**
* Fungsi fractional_knapsack menerima daftar item dan kapasitas ransel. Setiap
item direpresentasikan sebagai tuple (nilai, berat). Daftar tersebut diurutkan
berdasarkan rasio nilai per berat secara menurun. Fungsi kemudian
menambahkan barang sebanyak mungkin ke dalam ransel, dan jika tersisa
ruang tetapi tidak cukup untuk mengambil satu barang utuh, maka hanya
sebagian dari barang tersebut yang diambil sesuai sisa kapasitas. Total nilai
yang dikumpulkan dikembalikan sebagai hasil.

# Case 3: Minimum Spanning Tree: Prim dan Kruskal

Masalah **Minimum Spanning Tree (MST)** adalah salah satu aplikasi algoritma greedy
dalam konteks graf. Diberikan sebuah graf berbobot yang terhubung, tujuan dari MST
adalah menghubungkan semua simpul dalam graf dengan **total bobot minimum**,
**tanpa membentuk siklus**. Dua algoritma greedy yang populer untuk menyelesaikan
masalah ini adalah **Prim** dan **Kruskal**.

Kedua algoritma menggunakan prinsip greedy, tetapi dengan pendekatan yang
berbeda:
* **Algoritma Prim** memulai dari satu simpul dan secara bertahap menambahkan sisi
(edge) dengan bobot terkecil yang menghubungkan simpul yang sudah dikunjungi
dengan simpul yang belum.
* **Algoritma Kruskal** memulai dari semua sisi graf, lalu memilih sisi dengan bobot
terkecil yang tidak membentuk siklus, hingga semua simpul terhubung.
Kedua pendekatan ini menunjukkan bagaimana strategi greedy digunakan dalam
membangun solusi secara bertahap, dengan selalu memilih opsi lokal terbaik pada
tiap langkah.

**Contoh Kode Program 1:**

```python
import heapq

def prim_mst (graph, start):
  visited = set ()
  min_heap = [(0, start)]
  total_weight = 0

  while min_heap:
    weight, node = heapq.heappop (min_heap)
    if node in visited:
      continue

    visited.add (node)
    total_weight += weight
    for neighbor, edge_weight in graph[node] :
      if neighbor not in visited:
        heapq.heappush (min_heap, (edge_weight, neighbor))
  return total_weight

# Representasi graf: adjacency list
graph = {
'A': [('B', 2), ('C', 3)],
'B': [('A', 2), ('C', 1), ('D', 1)],
'C': [('A', 3), ('B', 1), ('D', 4)],
'D': [('B', 1), ('C', 4)]
}

print ("Total bobot MST (Prim) :", prim_mst (graph, 'A'))
```

**Contoh Kode Program 2:**

```python
def kruskal_mst (edges, n_nodes):
  parent = {i: i for i in range(n_nodes) }

  def find(x):
    while parent[x] != x:
      x = parent [x]
    return x

  def union (x, y):
    root_x = find (x)
    root_y = find (y)
    if root_x != root_y:
      parent[root_y] = root_x
      return True
    return False

  edges.sort (key=lambda x: x[2])
  total_weight = 0
  for u, v, weight in edges:
    if union(u, v):
      total_weight += weight
  return total_weight

# edges: (node1, node2, weight), nodes are represented by integers
edges = [(0, 1, 2), (0, 2, 3), (1, 2, 1), (1, 3, 1), (2, 3, 4)]
print("Total bobot MST (Kruskal):", kruskal_mst (edges, 4))
```

**Penjelasan:**
* Prim: Menggunakan priority queue (heapq) untuk memilih sisi terkecil yang
menghubungkan simpul yang sudah dikunjungi ke simpul lain. Hasilnya adalah
total bobot dari MST.
* Kruskal: Mengurutkan semua sisi berdasarkan bobot, lalu menggunakan union-
find (disjoint set) untuk menambahkan sisi satu per satu ke MST jika tidak
membentuk siklus. Total bobot dari semua sisi yang dipilih dijumlahkan.
