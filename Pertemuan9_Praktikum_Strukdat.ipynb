**Implementas Hash Table Sederhana**

```python
class SimpleHashTable:
  def __init__(self, size):
    self.size = size
    self.table = [None] * size

  def hash_function (self, key) :
    return hash (key) % self.size

  def insert(self, key, value) :
    index = self.hash_function (key)
    self.table[index] = value

  def search(self, key):
    index = self.hash_function (key)
    return self.table [index]

# Contoh penggunaan
ht = SimpleHashTable (10)
ht.insert ("apel", 100)
ht.insert("pisang", 200)

print ("Nilai untuk 'apel':", ht.search("apel"))
print ("Nilai untuk 'pisang':", ht.search("pisang"))
```

Penjelasan:
* Kelas SimpleHashTable memiliki array (self.table) untuk menyimpan data.
* Fungsi hash_function menggunakan fungsi hash bawaan Python, lalu
mengambil sisa bagi dengan ukuran tabel agar hasilnya sesuai dengan indeks
array.
* Metode insert memasukkan value ke indeks yang ditentukan oleh fungsi hash
dari key.
* Metode search mengakses data berdasarkan key yang di-hash.

Catatan: Karena belum menggunakan penanganan collision, jika dua key memiliki
hasil hash yang sama, data sebelumnya akan tertimpa.

**Penanganan Collision dengan Linear Probing**

```python
class LinearProbingHashTable:
  def __init__(self, size):
    self.size = size
    self.table = [None] * size

  def hash_function (self, key) :
    return hash(key) % self.size

  def insert(self, key, value):
    index = self.hash_function (key)
    original_index = index

    while self.table[index] is not None:
      if self.table[index] [0] == key:
        break
      index = (index + 1) % self.size
      if index == original_index:
        raise Exception ("Hash table penuh")

    self.table[index] = (key, value)

  def search(self, key):
    index = self.hash_function (key)
    original_index = index

    while self. table[index] is not None:
      if self.table[index] [0] == key:
        return self.table[index] [1]
      index = (index + 1) % self.size
      if index == original_index:
        break
    return None

# Contoh penggunaan
ht = LinearProbingHashTable (5)
ht.insert("apel", 100)
ht.insert("pisang", 200)
ht.insert ("melon", 300)

print ("Nilai untuk 'apel':", ht.search("apel"))
print ("Nilai untuk 'melon':", ht.search ("melon") )
```

Penjelasan:
* Jika terjadi tabrakan saat menyisipkan, insert() akan mencari indeks
berikutnya hingga menemukan slot kosong.
* Pencarian dalam search() juga menggunakan logika yang sama: lanjut ke
indeks berikutnya hingga menemukan key yang dicari, atau hingga seluruh
tabel diputar satu putaran penuh.
* Dengan linear probing, semua data tetap berada dalam array, tetapi penting
untuk menangani kemungkinan kondisi tabel penuh.

**Penanganan Collision dengan Chaining**

```python
class ChainingHashTable:
  def __init__(self, size) :
    self.size = size
    self.table = [[] for _ in range(size) ]

  def hash_function (self, key) :
    return hash(key) % self.size

  def insert(self, key, value) :
    index = self.hash_function (key)
    for i, (k, v) in enumerate(self.table[index]) :
      if k == key:
        self.table[index] [i] = (key, value)
        return
    self.table[index].append( (key, value))

  def search(self, key):
    index = self.hash_function(key)
    for k, v in self.table[index] :
      if k == key:
        return v
    return None

# Contoh penggunaan
ht = ChainingHashTable(5)
ht.insert ("apel", 100)
ht.insert("pisang", 200)
ht.insert ("melon", 300)

print ("Nilai untuk 'apel':", ht.search ("apel"))
print ("Nilai untuk 'melon':", ht.search ("melon"))
```

Penjelasan:
* self.table adalah array yang setiap elemennya berupa list kosong.
* Ketika terjadi collision, data tidak ditimpa atau dipindah, tapi ditambahkan ke
list di slot tersebut.
* Fungsi insert() juga melakukan pengecekan apakah key sudah ada
sebelumnya-jika ya, nilainya diperbarui.
* search() menelusuri list pada slot indeks tertentu untuk menemukan pasangan
key yang sesuai.

**1. Eksperimen Sederhana Linear Probing**

Modifikasi kode LinearProbingHashTable dengan menambahkan 5 data ke dalam tabel
yang berukuran 5.

```python
class LinearProbingHashTable:
    def __init__(self, size):
        self.size = size
        self.table = [None] * size

    def hash_function(self, key):
        return ord(key[0]) % self.size

    def insert(self, key, value):
        index = self.hash_function(key)

        # Linear Probing
        while self.table[index] is not None:
            index = (index + 1) % self.size

        self.table[index] = (key, value)


# Membuat objek hash table ukuran 5
ht = LinearProbingHashTable(5)

# Menambahkan data
ht.insert("A", 10)
ht.insert("B", 20)
ht.insert("C", 30)
ht.insert("D", 40)
ht.insert("E", 50)

# Mencetak isi tabel
print(ht.table)
```

Penjelasan

Linear Probing adalah metode penanganan collision pada hash table dengan cara mencari slot kosong berikutnya secara berurutan.

Jika dua key memiliki hasil hash yang sama, maka terjadi collision.
Sistem akan:

1. Mengecek apakah slot tujuan sudah terisi.
2. Jika penuh, pindah ke index berikutnya.
3. Proses dilakukan terus sampai menemukan slot kosong.

Contoh:

* Misalnya key "A" dan "F" menghasilkan index hash yang sama.
* "A" masuk lebih dulu.
Saat "F" dimasukkan, slot tersebut sudah terisi.
* Maka "F" ditempatkan di slot berikutnya.

Metode ini disebut Linear Probing karena pencarian dilakukan secara linear (berurutan).

**2. Percobaan Dasar Chaining**

Gunakan ChainingHashTable untuk menyisipkan beberapa data yang menghasilkan
collision.

```python
class ChainingHashTable:
    def __init__(self, size):
        self.size = size
        self.table = [[] for _ in range(size)]

    def hash_function(self, key):
        return len(key) % self.size

    def insert(self, key, value):
        index = self.hash_function(key)
        self.table[index].append((key, value))


# Membuat hash table ukuran 3
ht = ChainingHashTable(3)

# Menambahkan data
ht.insert("apel", 100)
ht.insert("melon", 200)
ht.insert("lemon", 300)
ht.insert("pisang", 400)

# Menampilkan isi tabel
for i, item in enumerate(ht.table):
    print(f"Index {i}: {item}")
```

Penjelasan

Pada metode Chaining, setiap slot pada hash table menyimpan list (rantai data).

Jika terjadi collision:

* Data baru tidak dipindahkan ke slot lain.
* Data langsung ditambahkan ke list pada index yang sama.

Contoh:

* "melon" dan "lemon" memiliki hasil hash yang sama.
* Keduanya disimpan dalam satu slot yang sama.
* Slot tersebut berisi lebih dari satu data dalam bentuk list.

Keuntungan chaining:

* Lebih mudah menangani banyak collision.
* Tidak perlu mencari slot kosong seperti pada linear probing.
