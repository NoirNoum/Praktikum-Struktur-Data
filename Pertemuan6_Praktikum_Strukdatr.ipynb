**IMPLEMENTASI BUBBLE SORT**

Contoh pengimplementasiannya:

```python
def bubble_sort (arr) :
  n = len (arr)
  for i in range(n - 1) :
    for j in range(n - i - 1):
      if arr[j] > arr[j + 1]:
        arr[j], arr[j + 1] - arr[j + 1], arr[j]
  return arr

# Contoh penggunaan
data = [64, 34, 25, 12, 22, 11, 90]
sorted_data = bubble_sort (data. copy () )  # Menggunakan copy agar data asli tidak berubah
print ("Data sebelum sorting:", data)
print ("Data setelah Bubble Sort:", sorted_data)
```

**Penjelasan:**
* Fungsi bubble_sort(arr) menerima sebuah daftar (array) sebagai input dan
mengurutkannya menggunakan metode Bubble Sort.
* Dua perulangan for bersarang digunakan untuk membandingkan elemen
secara berpasangan dan menukar jika perlu.
* arr[j] > arr[j + 1] memastikan elemen yang lebih besar berpindah ke posisi
yang lebih tinggi, sehingga elemen terbesar "menggelembung" ke posisi
terakhir dalam setiap iterasi.
* Hasil akhir adalah array yang terurut dari terkecil hingga terbesar.

Bubble Sort memiliki implementasi yang sederhana, tetapi tidak efisien untuk
jumlah data besar karena kompleksitas waktu O(n2). Selanjutnya, kita akan
mempelajari Selection Sort, yang menggunakan pendekatan berbeda dengan
mencari elemen terkecil dalam setiap iterasi.



---
**IMPLEMENTASI SELECTION SORT**

Contoh kode program:

```python
def selection_sort (arr) :
  n = len (arr)
  for i in range(n - 1) :
    min_index = i
    for j in range(i + 1, n) :
      if arr[j] < arr[min_index] :
        min_index = j
        arr[i], arr[min_index] - arr[min_index], arr[i]
  return arr

# Contoh penggunaan
data = [64, 25, 12, 22, 11]
sorted_data = selection_sort (data. copy () )
print ("Data sebelum sorting:", data)
print ("Data setelah Selection Sort:", sorted_data)
```

**Penjelasan:**
* Fungsi selection_sort(arr) menerima sebuah daftar dan mengurutkannya
menggunakan metode Selection Sort.
* Variabel min_index digunakan untuk menyimpan indeks elemen terkecil
dalam iterasi saat ini.
* Perulangan for j mencari elemen terkecil dari sisa daftar dan memperbarui
min_index.
* Setelah indeks terkecil ditemukan, elemen pada posisi i dan min_index
ditukar, sehingga elemen terkecil berada di posisi yang benar.

Selection Sort lebih efisien dibandingkan Bubble Sort dalam hal jumlah pertukaran,
tetapi masih memiliki kompleksitas waktu O(n2). Untuk meningkatkan efisiensi, kita
bisa menggunakan algoritma Insertion Sort, yang akan kita bahas pada subbab
berikutnya.



---
**IMPLEMENTATION INSERTION SORT**

Contoh kode program:

```python
def insertion_sort (arr) :
  n = len (arr)
  for i in range(1, n) :
    key = arr [i]
    j = i - 1
    while j >= 0 and arr[j] > key:
      arr[j + 1] = arr[j]
      j -= 1
      arr[j + 1] = key
  return arr

# Contoh penggunaan
data = [64, 34, 25, 12, 22, 11, 90]
sorted_data = insertion_sort (data. copy () )
print ("Data sebelum sorting:", data)
print ("Data setelah Insertion Sort:", sorted_data)
```

**Penjelasan:**
* Fungsi insertion_sort(arr) menerima daftar dan mengurutkannya menggunakan
metode Insertion Sort.
* Variabel key menyimpan nilai elemen yang akan dimasukkan ke dalam posisi
yang benar.
* Perulangan while digunakan untuk menggeser elemen-elemen yang lebih besar
ke kanan sampai ditemukan posisi yang sesuai untuk key.
* Setelah ditemukan posisi yang benar, nilai key dimasukkan ke dalam daftar.

Insertion Sort lebih efisien dibandingkan Bubble Sort dan Selection Sort untuk
dataset kecil atau hampir terurut. Namun, untuk dataset besar, kita membutuhkan
algoritma yang lebih cepat seperti Merge Sort, yang akan kita bahas pada subbab
berikutnya.

IMPLEMENTATION MERGE SORT


Contoh kodenya:

```python
def merge_sort(arr):
  if len(arr) > 1:
    mid = len (arr) // 2
    left_half = arr [:mid]
    right_half = arr[mid:]

    merge_sort(left_half)
    merge_sort(right_half)

    i = j = k = 0

    while i < len (left_half) and j < len (right_half) :
      if left_half[i] < right_half[j]:
        arr[k] = left_half[i]
        i += 1
      else:
        arr[k] - right_half [j]
        j +- 1
        k +- 1

    while i < len(left_half) :
      arr[k] = left_half[i]
      i += 1
      k +- 1

    while j < len (right_half) :
      arr[k] = right_half[j]
      j += 1
      k += 1

  return arr

# Contoh penggunaan
data = [64, 25, 12, 22, 11]
sorted_data = merge_sort (data. copy () )
print ("Data sebelum sorting:", data)
print ("Data setelah Merge Sort:", sorted_data)
```

Penjelasan:
* Fungsi merge_sort(arr) membagi daftar menjadi dua bagian (left_half dan
right_half) hingga setiap bagian hanya memiliki satu elemen.
* Rekursi digunakan untuk mengurutkan masing-masing bagian sebelum
digabungkan kembali.
* Proses penggabungan (merging) dilakukan dengan membandingkan
elemen-elemen dari kedua bagian dan memasukkannya ke dalam daftar utama
(arr).
* Dua perulangan while terakhir memastikan semua elemen tersisa dimasukkan ke
dalam daftar utama jika masih ada yang belum tergabung.

Merge Sort menawarkan efisiensi yang lebih tinggi dibandingkan algoritma sorting
sebelumnya, terutama untuk dataset besar. Namun, ada algoritma lain yang juga
berbasis divide and conquer dan sering digunakan dalam dunia nyata, yaitu Quick
Sort, yang akan kita bahas pada subbab berikutnya.



---
**IMPLEMENTATION QUICK SORT**


Contoh kode program:

```python
def quick_sort(arr):
  if len(arr) <= 1:
    return arr

  pivot = arr[len(arr) // 2]
  left = [x for x in arr if x < pivot]
  middle = [x for x in arr if x == pivot]
  right = [x for x in arr if x > pivot]

  return quick_sort (left) + middle + quick_sort (right)

# Contoh penggunaan
data = [64, 25, 12, 22, 11]
sorted_data = quick_sort (data.copy())
print ("Data sebelum sorting:", data)
print ("Data setelah Quick Sort:", sorted_data)
```

**Penjelasan:**
* Fungsi quick_sort(arr) bekerja secara rekursif dengan memilih pivot sebagai titik
pembanding.
* Daftar dibagi menjadi tiga bagian:
    * left : elemen yang lebih kecil dari pivot
    * middle: elemen yang sama dengan pivot
    * right: elemen yang lebih besar dari pivot
* Proses rekursi dilakukan pada left dan right hingga seluruh elemen terurut.
* Akhirnya, semua bagian digabungkan kembali menjadi satu daftar yang sudah
terurut.

Quick Sort adalah salah satu algoritma sorting tercepat dan sering digunakan dalam
berbagai aplikasi nyata.

**Rekomendasi Film atau Lagu berdasarkan Rating (Rombel 04)**

Platform seperti Netflix atau Spotify biasanya menampilkan:

* Film dengan rating tertinggi
* Lagu paling populer

Aku akan mengurutkan data berdasarkan rating (descending)

```python
def quick_sort_pilem(arr):
    if len(arr) <= 1:
        return arr

    pivot = arr[len(arr)//2]["rating"]

    left = [x for x in arr if x["rating"] > pivot]   # lebih besar dulu (descending)
    middle = [x for x in arr if x["rating"] == pivot]
    right = [x for x in arr if x["rating"] < pivot]

    return quick_sort_pilem(left) + middle + quick_sort_pilem(right)

# Data film
movie = [
    {"judul": "Avengers Sevenvold", "rating": 8.5},
    {"judul": "Interstellar", "rating": 9.0},
    {"judul": "Badut Baik yang Tersakiti", "rating": 8.4},
    {"judul": "Pengabdi Bahlil", "rating": 7.5},
    {"judul": "", "rating": 8.8}
]

# Sorting
sorted_movie = quick_sort_pilem(movie)

# Output
print("Daftar Film Berdasarkan Rating Tertinggi:")
for movie in sorted_movie:
    print(movie["judul"], "-", movie["rating"])
```
